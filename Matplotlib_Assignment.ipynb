```python
''' 
Question 2 Ans 

    A scatter plot is a type of chart used to display the relationship between two variables. Each point
on the plot represents a pair of values for the two variables being plotted, and the position of the point
indicates the values for each variable.
    Scatter plots are commonly used in data analysis to identify patterns or relationships between two
variables. For example, a scatter plot could be used to visualize the relationship between a person's height
and weight, or the relationship between a company's revenue and profit. In a scatter plot, the x-axis 
represents one variable, and the y-axis represents the other variable. The data points are then plotted based
on their values for each variable. The points on the scatter plot can be color-coded or labeled to indicate
additional information about the data, such as the category or group to which each point belongs.

    Scatter plots are useful for identifying trends and outliers in the data. A trend is a pattern that can be
observed in the data, such as a general increase or decrease in one variable as the other variable increases.
An outlier is a point that is far away from the other points on the plot, and may represent a data entry error
or an unusual observation.

    Matplotlib's Pyplot module provides a simple and intuitive interface for creating scatter plots in Python.
The scatter() function can be used to create a scatter plot with the x and y values as arguments, and additional
parameters can be used to customize the appearance of the plot, such as the color and size of the points.
'''
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
'''
    The subplot() function in Matplotlib is used to create multiple plots within the same figure.
It allows you to create a grid of plots, where each plot displays a different aspect of the data
or a different set of data altogether.

    The subplot() function takes three arguments: the number of rows in the grid, the number of
columns in the grid, and the index of the subplot to create. The index starts from 1 and goes from
left to right and top to bottom in the grid.

'''
```

```python

# Line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of plots
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title("Line Plot 1")

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title("Line Plot 2")

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title("Line Plot 3")

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title("Line Plot 4")

# Add a main title and display the plot
plt.suptitle("Multiple Line Plots")
plt.show()
```

```python
'''
    A bar plot, also known as a bar chart, is a graphical representation of data in which rectangular bars
are used to display the values of a categorical variable or a group of categorical variables. The length of
each bar is proportional to the value it represents.
    Bar plots are commonly used to compare and display the relative frequencies, counts, or other numerical
summaries of different categories. They are often used to visualize the distribution of categorical data, as
well as to identify patterns, trends, and relationships between different groups.
    One of the advantages of bar plots is that they are easy to read and interpret. They are also versatile 
and can be used to display data in different ways, such as stacked bar plots, grouped bar plots, or horizontal
bar plots.
    Overall, bar plots are a useful tool for summarizing and visualizing categorical data, making it easier to
    understand and interpret complex information
    
'''
```

```python

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit (in millions)")
plt.title("Company profits")
plt.show()
 
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.barh(company, profit)
plt.ylabel("Company")
plt.xlabel("Profit (in millions)")
plt.title("Company profits")
plt.show()
```

```python
'''
    A box plot, also known as a box-and-whisker plot, is a graphical representation of a distribution of numerical data through their quartiles.
The box in the plot shows the interquartile range (IQR), which is the range between the first quartile (Q1) and the third quartile (Q3) of the
data. The median (Q2) is represented by a line inside the box. The whiskers extend from the box to the minimum and maximum values that are within
1.5 times the IQR from the Q1 and Q3, respectively, and any points beyond the whiskers are considered outliers.
    Box plots are useful for visualizing the spread and skewness of a dataset, as well as identifying potential outliers. They are commonly used
in data analysis and statistics to compare the distributions of different groups or variables, or to visualize the distribution of a single variable.
'''
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.ylabel("Values")
plt.title("Box Plot")
plt.show()
```
