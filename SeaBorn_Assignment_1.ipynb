```python
''' 
Scatter Plot -
    A scatter plot is used to display the relationship between two variables.
    It is useful for identifying patterns and trends in the data.

Line Plot -
    A line plot is used to show how the data changes over time.
    It can be used to identify trends or to compare different groups of data.

Histogram -
    A histogram is used to display the distribution of a single variable.
    It can be used to identify the shape of the distribution, the presence of outliers,
    or to compare the distribution of different groups.

Box Plot -
    A box plot is used to display the distribution of a single variable or the relationship
    between two variables. It can be used to identify the median, quartiles, outliers, and
    the range of the data.

Heatmap -
    A heatmap is used to display the relationship between two variables using colors.
    It is useful for identifying patterns in large datasets and for visualizing correlations 
    between variables.
'''
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fmri = sns.load_dataset("fmri")
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
sns.set_style("darkgrid")
sns.set(rc={'figure.figsize':(10,8)})
sns.set(font_scale=1.2)
plt.title("Signal over Time by Event and Region")
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
sns.boxplot(x='pclass', y='age', data=titanic)
sns.boxplot(x='pclass', y='fare', data=titanic)
sns.set_style("darkgrid")
sns.set(rc={'figure.figsize':(10,8)})
sns.set(font_scale=1.2)
plt.show()
```

```python
diamonds = sns.load_dataset('diamonds')
sns.histplot(data=diamonds, x='price', hue='cut')
sns.set_style("darkgrid")
sns.set(rc={'figure.figsize':(10,8)})
sns.set(font_scale=1.2)
plt.title('Histogram of Diamond Prices by Cut')
plt.show()
```

```python
flights = sns.load_dataset('flights')
flights = flights.pivot('month', 'year', 'passengers')
sns.heatmap(flights, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Number of Passengers (1000s)')
plt.show()
```
